#### filtrando una region

```python
from shapely.geometry import box
```

```python
#Definimos cuadrante de regiones
min_long = -43.24
min_lat = -23.00
max_long = -43.13
max_lat = -22.88

# Crear un polígono de la caja delimitadora
bbox_polygon = box(min_long, min_lat, max_long, max_lat)
```

```python
gdf_filtered = gdf1[gdf1.geometry.intersects(bbox_polygon)]
```

```python
gdf_filtered
```

```python
gdf_filtered.plot()
plt.title("Bairros RJ City Filtered")
plt.show()
```

```python
output_geojson_file = 'Bairros_participants_rj.geojson'
gdf_filtered.to_file(output_geojson_file, driver='GeoJSON')
```
