```python
import pandas as pd
import ast
import json
```

```python
df_scores = pd.read_excel("Puntuacion.xlsx", engine="openpyxl")
p_cols = [f'P{i}' for i in range(1, 51)]
df_scores['scores'] = df_scores[p_cols].values.tolist()
df_scores.drop(columns=p_cols, inplace=True)
df_scores
```

```python
df_lista = pd.read_excel("listaAlet150.xlsx")
def to_int_array(s):
    s = s.replace("{", "[").replace("}", "]")  
    lista_str = ast.literal_eval(s)           
    return [int(x) for x in lista_str]         

df_lista["seleccionadas"] = df_lista["seleccionadas"].apply(to_int_array)
df_lista
```

```python
only_number_participante = list(df_lista['participante_id'])
new_column = []
for i in only_number_participante:
    new_column.append(int(i[13:]))

df_lista['number_participant'] = new_column
diagram = df_lista[['number_participant', 'seleccionadas']]
diagram
```

```python
result_json = diagram.set_index('number_participant')['seleccionadas'].apply(list).to_dict()

with open('/home/sante/Documents/FGV/Eclass/Data Visualization/final-project-security-urban-perception-tracking/static/diagram.json', 'w') as f:
    json.dump(result_json, f, indent=2)
```

```python
df = pd.concat([df_scores, df_lista], axis=1)
df = df[['Nº PARTICIPANTE', 'ESTADO/PAIS DE ORIGEM', 'BAIRRO', 'IDADE', 'GRAU DE ESCOLARIDADE', 'IDENTIDADE DE GÊNERO', 'seleccionadas', 'scores']]
df = df.rename(columns={
    'score': 'Score Safety',
    'seleccionadas': 'Id Image'
})
df.head()
```

```python
cross = {}
for i in range(len(df['Id Image'])):
    for j in range(len(df['Id Image'][i])):
        key = df['Id Image'][i][j]
        if key not in cross:
            cross[key] = []
        
        cross[key].append({
            'score': int(df['scores'][i][j]),
            'participant': int(df['Nº PARTICIPANTE'][i]),
            'state': df['ESTADO/PAIS DE ORIGEM'][i],
            'age': int(df['IDADE'][i]),
            'gener': df['IDENTIDADE DE GÊNERO'][i],
        })
```

```python
from collections import OrderedDict

data_sorted = OrderedDict(sorted(cross.items(), key=lambda x: int(x[0])))

avg_scores = {}
for key, records in data_sorted.items():
    if records: 
        total_score = sum(item['score'] for item in records)
        avg_score = total_score / len(records)
        avg_scores[key] = round(avg_score, 5)  
    else:
        avg_scores[key] = None
```

```python
df = pd.DataFrame(list(avg_scores.items()), columns=['id', 'avg_score'])
```

```python
merge = pd.read_csv("temp.csv")
merge["avg_hololens"] = df["avg_score"]
merge["avg_pp2"] = merge["safety"]
merge["avg_pp2"] = merge["avg_pp2"].round(1)
merge = merge.drop(columns=['safety']) 
```

```python
merge.to_json("image_safety.json", orient="records", indent=2)
with open("image_safety.json", "r") as f:
    data_safety = json.load(f)

combined_data = {}
for i, (safety_data, sorted_key) in enumerate(zip(data_safety, data_sorted)):
    key = str(i)
    combined_data[key] = {
        **safety_data,
        "score_participant": data_sorted[sorted_key]
    }

with open("/home/sante/Documents/FGV/Eclass/Data Visualization/final-project-security-urban-perception-tracking/static/data_hololens.json", "w") as f:
    json.dump(combined_data, f, indent=2, ensure_ascii=False)
```
